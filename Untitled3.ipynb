```python
import requests

class SecurityAuditor:
    def __init__(self, target_url):
        # الموقع الافتراضي المخصص للاختبار القانوني
        self.target_url = target_url 
        self.headers = {
            'User-Agent': 'Ethical-Security-Researcher-Student'
        }

    def run_audit(self):
        print(f"--- بدء الفحص القانوني على: {self.target_url} ---")
        
        # 1. فحص الرؤوس الأمنية
        try:
            response = requests.get(self.target_url, headers=self.headers, timeout=5)
            headers_to_check = ['Content-Security-Policy', 'X-Frame-Options', 'Strict-Transport-Security']
            
            print("\n[+] نتائج فحص الـ Headers:")
            for h in headers_to_check:
                status = "✅ موجود" if h in response.headers else "🔴 مفقود"
                print(f"  - {h}: {status}")
        except Exception as e:
            print(f"[-] خطأ في الاتصال: {e}")

        # 2. فحص أمان الشبكة (مستقل عن الموقع)
        print("\n[+] فحص خصوصية الاتصال (Network Privacy):")
        res_privacy = requests.get("https://httpbin.org/headers")
        if "msisdn" in res_privacy.text.lower():
            print("  [⚠️] تنبيه: بيانات الاتصال مسربة!")
        else:
            print("  [✅] الاتصال آمن والبيانات الشخصية محجوبة.")

# استخدام موقع افتراضي (قانوني 100% للفحص)
# هذا الموقع صُمم خصيصاً ليتم "اختراقه" وفحصه من قبل الطلاب
test_auditor = SecurityAuditor("http://testphp.vulnweb.com")
test_auditor.run_audit()
```
